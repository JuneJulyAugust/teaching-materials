# Fast Fourier Transform

A fast Fourier transform (FFT) is an algorithm to compute the discrete Fourier transform in a rapidly way.

FFTs are widely used in engineering, science and mathematics.

A FFT computes the DFT and produces exactly the same result as evaluating the DFT definition directly, but is much faster.

### Discrete Fourier transform (DFT)

The DFT of an input vector $x$ of length $N$ is defined as the matrix-vector product:

$$ y = \text{DFT}_N x $$

where

$$ \text{DFT}_N = \left[w_N^{kl}\right]_{0\leq k,l < N} $$

with $w_N = \exp\left(-2\pi/N\right)$

Evaluating this definition reauires $O(N^2)$ operations: there are $N$ outputs and each one requires a sum of $N$ terms.

An FFT is any method to compute the same results in $O(N\log N)$ operations.

### Kronecker product formalism

The Kronecker product of matrices $A$ and $B$ is defined as

$$ A \otimes B = \left[ a_{k,l}B \right]_{k,l} $$

If $I_n$ represents the $n\times n$Identity matrix, we get

$$ I_n \otimes A =
\begin{bmatrix}
A &   &        &   \\
  & A &        &   \\
  &   & \ddots &   \\
  &   &        & A
\end{bmatrix} $$

a $n$-block diagonal matrix where each block is the matrix $A$.

<center>![image](files/images/kronecker_product.png)</center>

We can also introduce the iterative direct sum:

$$ \bigoplus_{i=0}^{n-1} A_i =  
\begin{bmatrix}
A_0 &   &        &   \\
  & A_1 &        &   \\
  &   & \ddots &   \\
  &   &        & A_{n-1}
\end{bmatrix} $$

The Kronecker product is not commutative. $A \otimes I_n$ contains also $n$ blocks of $A$, but are spread out and intervaled at stride $n$. 

<center>![image](files/images/kronecker_product_2.png)</center>

All elements with the same shade of gray taken together constitute one $A$. So the final matrix contains also four $A$ matrices.

````python
import numpy

N = 16    # size of input vector
m = 4     # stride
n = N/m   # ratio

x_input = numpy.arange(N, dtype=int)    # input vector

def permutation(x_input, m, n):
    """Performs permutation at stride m.
    
    Arguments
    ---------
    x_input -- input vector.
    m -- stride.
    n -- ratio between the input size and the stride.
    
    Returns
    -------
    x_ouput -- the permuted output vector.
    """
    x_output = numpy.empty_like(x_input)
    
    for i in xrange(m):
        for j in xrange(n):
            x_output[i+m*j] = x_input[n*i+j]
    
    return x_output

# perform permutation
x_output = permutation(x_input, m, n)

# output results
print 'x_input =\n', x_input, '\n'
print 'x_output =\n', x_output, '\n'
````

### Cooley-Tukey algorithm (1965)

This is by far the most commonly used FFT! The algorithm recursively breaks down a DFT of any size $N = mn$ into smaller DFTs of sizes $m$ and $n$, along with $O(N)$ multiplications by complex roots of unity traditionally called twiddle factors.

Using the Kronecker product formalism, the recursive Cooley-Tukey algorithm is given by the following expression:

$$ \text{DFT}_{mn} = \left( \text{DFT}_m \otimes I_n \right) D_{m,n} \left( I_m \otimes \text{DFT}_n \right) L_m^{mn} $$

where

$$ D_{m,n} = \bigoplus_{j=0}^{m-1} \text{diag}\left( \left[w_{mn}^i\right]_{0\leq i < n} \right)^j $$

is called the twiddle matrix.

The discrete Fourier transform can now be computed in four steps:

* the input vector $x$ is permuted by $L_m^{mn}$,
* $m$ DFTs of size $n$ are computed recursively on segments of the input vector $x$,
* the vector is scaled element-wise by the twiddle matrix $D_{m,n}$,
* $n$ DFTs of size $m$ are computed recursively at stride $m$.

Observation: The multiplication of a vector by a Kronecker product containing an identity matrix can be computed using loops.

The working set for each of the $m$ iterations of $y=\left(I_m \otimes A_n \right) x$ is a contiguous block of size $n$ and the base address is increased by $n$ between iterations.

````python
def Im_x_An(x, An, n, m):
    """Performs the Kronecker product between identity matrix 
    and a square matrix.
    
    Arguments
    ---------
    x -- input vector of length nxm.
    An -- nxn square matrix.
    n,m -- sizes of the matrices.
    
    Returns
    -------
    y -- output of the Kronecker product.
    """
    y = numpy.empty_like(x)
    
    # m iterations
    for i in xrange(0, m, 1):
        y[i*n:i*(n+1):1] = An.dot(x[i*n:i*(n+1):1])
    
    return y    
````

In contrast, the working sets of size $m$ of the $n$ iterations of $y = \left( A_m \otimes I_n \right) x$ are intervaled, leading to stride $n$ within one iteration.

````python
def Am_x_In(x, Am, n, m):
    """Performs the Kronecker product between a square matrix
    and the identity matrix .
    
    Arguments
    ---------
    x -- input vector of length nxm.
    Am -- mxm square matrix.
    n,m -- sizes of the matrices.
    
    Returns
    -------
    y -- output of the Kronecker product.
    """
    y = numpy.empty_like(x)
    
    # m iterations
    for i in xrange(0, m, 1):
        y[i:i+m:n] = Am.dot(x[i:i+m:n])
    
    return y
````

### Cache optimization

#### Blocking

We choose a radix-$4$ implementation of the Cooley-Tukey algorithm with $N=4^n$, which means that $N$ is factorized as follow: $N = 4 . 4^{n-1}$. The radix $m=4$ is chosen as the value for the block size.

Therefore, the discrete Fourier transform of an input vector $x$ of length $N=4^n$ is recursively computed using:

$$ \text{DTF}_{4^n} = \left( \text{DTF}_4 \otimes I_{4^{n-1}} \right) D_{4,4^{n-1}} \left( I_4 \otimes \text{DTF}_{4^{n-1}} \right) L_4^{4^n} $$

The terms $\text{DTF}_4$ are the recursion leaves.

Let's have a look at the recursion for $n=2$:

<center>![image](files/images/dft16.png)</center>

#### Loop merging

A naive implementation leads to a recursive function with four stages and thus four sweeps through the data. However, the stride permutation $L_4^{4^n}$ is just a data reordering and thus is a candidate for loop merging. Similarly, the twidle factor matrix $D_{4,4^{n-1}}$ is a diagonal matrix and can be merged with the subsequent stage.

The recursion can be cast into two stages (instead of four):

$$ \text{DTF}_{4^n} = \left( \left( \text{DTF}_4 \otimes I_{4^{n-1}} \right) D_{4,4^{n-1}} \right) \left( \left( I_4 \otimes \text{DTF}_{4^{n-1}} \right)  L_4^{4^n} \right) $$

For $n=2$, this gives:
$$ \text{DTF}_{16} = \left( \left( \text{DTF}_4 \otimes I_4 \right) D_{4,4} \right) \left( \left( I_4 \otimes \text{DTF}_4 \right)  L_4^{16} \right) $$

<center>![image](files/images/dft16_merged.png)</center>

As stated before, the Kronecker product $I_4 \otimes \text{DTF}_{4^{n-1}}$ can be computed using a loop with $4$ iterations. The same is true with $\left( I_4 \otimes \text{DTF}_{4^{n-1}} \right) L_4^{4^{n-1}}$: the input is now read at stride $4$ but the output is still written at stride $1$.

Because the input is read at stride $4$, the radix, and the output is written at stride $1$, $x$ and $y$ need to be different memory regions.

The prototype of the recursive function would be:

```cpp
void DFT_rec(int N, int n, float *Y, float *X, int s);
```

The function would be called at the top-level by:
```cpp
DTF_rec(N, log4(N), Y, X, 1)
```

For $N=4$, at the leaf of the recursion, the kernel `DTF4_base` is called, whose prototype is:
```cpp
void DTF4_base(float *Y, float *X, int s);
```

The second stage of the recursive function computes $\left( \text{DTF}_4 \otimes I_{4^{n-1}} \right) D_{4,4^{n-1}}$. It first scales the input by a diagonal matrix and then sweeps with a $\text{DTF}_4$ kernel over it, applied at a stride.

Those two steps can be merged into one. The $\text{DTF}_4$s can be replaced by $\text{DTF}_4D_j$, where $D_j$ is a $4 \times 4$ diagonal matrix containing the proper diagonal elements from $D_{4, 4^{n-1}}$.

$$D_j = \text{diag} \left( w_{4^n}^0, w_{4^n}^j, w_{4^n}^{2j}, w_{4^n}^{3j} \right) , \quad 0\leq j < 4^{n-1}$$

Hence, the function implementing $y=\left( \text{DTF}_4 D_j \right)x$ also needs a stride as parameter, and $j$ to compute the elements of $D_j$. Also, it can be in-place since its, reads from and writes to the same locations of input and output vector. The prototype of the function looks like:
```cpp
void DTF4_twiddle(float *Y, int s, int n, int j);
```

The source code would look like:

```cpp
#include <math.h>

#define log4(N) (int)(log(N)/log(4))

#define DFT(N, Y, X) DTF_rec(N, log4(N), Y, X, 1)

// prototypes
void DTF4_base(float *Y, float *X, int s);
void DTF4_twiddle(float *Y, int s, int n, int j);

void DTF_rec(int N, int n, float *Y, float *X, int s)
{
    if (N==4) // leaf of the recursion
    {
        DTF4_base(Y, X, s);
    }
    else // recursive call
    {
        // first stage
        for (int j=0; j<4; j++)
        {
            DTF_rec(N/4, n-1, Y+(N/4)*j, X+j*s, s*4);
        }
        // second stage
        for (int j=0; j<N/4; j++)
        {
            DTF4_twiddle(Y+j, N/4, n, j);
        }
    }
}
```

````python
def DTF_rec(N, n, Y, X, s):
    """Computes the discrete Fourier transform in a recursive way.
    
    Arguments
    ---------
    N -- length of the input vector (in powers of 4).
    n -- power in base 4 of the input length (N=4^n).
    Y -- output vector.
    X -- input vector.
    s -- stride.
    """
    # recursion leaf
    if N==4:
        DTF4_base(Y, X, s)
    else:
        # first stage, recursive call
        for j in xrange(4):
            DTF_rec(N/4, n-1, Y[N/4*j:], X[])
````

Buffering: The kernel `DTF4_twiddle` accesses both input and output in a stride. For large sizes $N=4^n$, this stride is large

```cpp
// cache line size of 64 bytes (16 floats)
#define LS 16

void DFT_buf_rec(int N, int n, float *Y, float *X, int s, int th)
{
    // local buffer
    float buf[4*LS];
    
    if (N==4) // recursion leaf
    {
        DTF4_base(Y, X, s);
    }
    else // recursive call
    {
        // first stage
        if (N>th)
        {
            for (int j=0; j<4; j++)
            {
                DFT_buf_rec(N/4, n-1, Y+(N/4)*j, X+j*s, s*4, th);
            }
        }
        else
        {
            for (int j=0; j<4; j++)
            {
                DFT_rec(N/4, n-1, Y+(N/4)*j, X+j*s, s*4);
            }
        }
        // second stage
        for (int j1=0; j1<N/(4*LS); j1++)
        {
            // copy 4 chunks of LS float to local buffer
            for (int i=0; i<4; i++)
            {
                for (int k=0; k<LS; k++)
                {
                    buf[LS*i+k] = Y[LS*j1+(N/4)*i+k];
                }
            }
            // perform LS DTF4 on contiguous data
            for (int j2=0; j2<LS; j2++)
            {
                DTF4_twiddle(buf+j2, LS, n, j1*LS+j2);
            }
            // copy 4 chunks of LS float to output
            for (int i=0; , i<4; i++)
            {
                for (int k=0; k<LS; k++)
                {
                    Y[LS*j1+(N/4)*i+k] = buf[LS*i+k]
                }
            }
        }
    }
}
```
